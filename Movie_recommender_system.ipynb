```python
#helper function
import ast
def func(x):
    y = []
    for i in ast.literal_eval(x):
        y.append(i['name'])
    return y
```

```python
def fun2(x):
    L = []
    for i in ast.literal_eval(x):
        L.append(i['name'])
    return L
```

```python
movies['keywords'] = movies['keywords'].apply(fun2)
```

```python
movies
```

```python
def fun3(x):
    L = []
    counter = 0
    for i in ast.literal_eval(x):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(fun3)
```

```python
movies.head(1)
```

```python
def fetch_director(x):
    L = []
    for i in ast.literal_eval(x):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
        
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

Text - vectors (bag of words)

calculate most 5000 used words
then calculate the no of times each word come in each movie.

5000X5000.
each movie is an vector.
we won't consider stop words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words = 'english' )
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

Stemming 

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem('loves')
```

```python
def stem(text):
    y = []
    
    for i in text.split(): #coverting string to list
        y.append(ps.stem(i))
    return " ".join(y) #converting back the list to string
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

Euclidean distance is not a good option for high dimnetional data, so we use cosine distance.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

movie to other movie

```python
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
